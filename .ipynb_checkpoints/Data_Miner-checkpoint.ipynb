## Daily cases and deaths by date reported to WHO

### Download link: https://covid19.who.int/WHO-COVID-19-global-data.csv

#### Field name	Type	Description
1. Date_reported	Date	Date of reporting to WHO
1. Country_code	String	ISO Alpha-2 country code
1. Country	String	Country, territory, area
1. WHO_region	String	WHO regional offices: WHO Member States are grouped into six WHO regions -- Regional Office for Africa (AFRO), Regional Office for the Americas (AMRO), Regional Office for South-East Asia (SEARO), Regional Office for Europe (EURO), Regional Office for the Eastern Mediterranean (EMRO), and Regional Office for the Western Pacific (WPRO).
1. New_cases	Integer	New confirmed cases. Calculated by subtracting previous cumulative case count from current cumulative cases count.*
1. Cumulative_cases	Integer	Cumulative confirmed cases reported to WHO to date.
1. New_deaths	Integer	New confirmed deaths. Calculated by subtracting previous cumulative deaths from current cumulative deaths.*
1. Cumulative_deaths	Integer	Cumulative confirmed deaths reported to WHO to date.

```python
# read url into df
WHO_COVID_19_global_data = pd.read_csv('https://covid19.who.int/WHO-COVID-19-global-data.csv')
# save df
WHO_COVID_19_global_data.to_csv('mined_csv_data/WHO_COVID_19_global_data.csv')
# inspect df
WHO_COVID_19_global_data.head(2)
```

## Latest reported counts of cases and deaths, and transmission classification

### Download link: https://covid19.who.int/WHO-COVID-19-global-table-data.csv

#### Field name	Type	Description
1. Name	String	Country, territory, area
1. WHO_region	String	WHO Region
1. Cases - cumulative total	Integer	Cumulative confirmed cases reported to WHO to date.
1. Cases - cumulative total per 100000 population	Decimal	Cumulative confirmed cases reported to WHO to date per 100,000 population.
1. Cases - newly reported in last 7 days	Integer	New confirmed cases reported in the last 7 days. Calculated by subtracting previous cumulative case count (8 days prior) from current cumulative cases count.
1. Cases - newly reported in last 7 days per 100000 population	Decimal	New confirmed cases reported in the last 7 days per 100,000 population.
1. Cases - newly reported in last 24 hours	Integer	New confirmed cases reported in the last 24 hours. Calculated by subtracting previous cumulative case count from current cumulative cases count.
1. Deaths - cumulative total	Integer	Cumulative confirmed deaths reported to WHO to date.
1. Deaths - cumulative total per 100000 population	Decimal	Cumulative confirmed deaths reported to WHO to date per 100,000 population.
1. Deaths - newly reported in last 7 days	Integer	New confirmed deaths reported in the last 7 days. Calculated by subtracting previous cumulative death count (8 days prior) from current cumulative deaths count.
1. Deaths - newly reported in last 7 days per 100000 population	Decimal	New confirmed deaths reported in the last 7 days per 100,000 population.
1. Deaths - newly reported in last 24 hours	Integer	New confirmed deaths reported in the last 24 hours. Calculated by subtracting previous cumulative death count from current cumulative deaths count.

```python
# read url into df
WHO_COVID_19_global_table_data = pd.read_csv('https://covid19.who.int/WHO-COVID-19-global-table-data.csv')
# save df
WHO_COVID_19_global_table_data.to_csv('mined_csv_data/WHO_COVID_19_global_table_data.csv')
# inspect df
WHO_COVID_19_global_table_data.head(2)
```

## Vaccination data

### Download links: https://covid19.who.int/who-data/vaccination-data.csv

#### Field name	Type	Description
1. COUNTRY	String	Country, territory, area
1. ISO3	String	ISO Alpha-3 country code
1. WHO_REGION	String	WHO regional offices: WHO Member States are grouped into six WHO regions: Regional Office for Africa (AFRO), Regional Office for the Americas (AMRO), Regional Office for South-East Asia (SEARO), Regional Office for Europe (EURO), Regional Office for the Eastern Mediterranean (EMRO), and Regional Office for the Western Pacific (WPRO).
1. DATA_SOURCE	String	Indicates data source: - REPORTING: Data reported by Member States, or sourced from official reports - OWID: Data sourced from Our World in Data: https://ourworldindata.org/covid-vaccinations
1. DATE_UPDATED	Date	Date of last update
1. TOTAL_VACCINATIONS	Integer	Cumulative total vaccine doses administered
1. PERSONS_VACCINATED_1PLUS_DOSE	Decimal	Cumulative number of persons vaccinated with at least one dose
1. TOTAL_VACCINATIONS_PER100	Integer	Cumulative total vaccine doses administered per 100 population
1. PERSONS_VACCINATED_1PLUS_DOSE_PER100	Decimal	Cumulative persons vaccinated with at least one dose per 100 population
1. PERSONS_FULLY_VACCINATED	Integer	Cumulative number of persons fully vaccinated
1. PERSONS_FULLY_VACCINATED_PER100	Decimal	Cumulative number of persons fully vaccinated per 100 population
1. VACCINES_USED	String	Combined short name of vaccine: “Company - Product name” (see below)
1. FIRST_VACCINE_DATE	Date	Date of first vaccinations. Equivalent to start/launch date of the first vaccine administered in a country.
1. NUMBER_VACCINES_TYPES_USED	Integer	Number of vaccine types used per country, territory, area

```python
# read url into df
vaccination_data = pd.read_csv('https://covid19.who.int/who-data/vaccination-data.csv')
# save df
vaccination_data.to_csv('mined_csv_data/vaccination_data.csv')
# inspect df
vaccination_data.head(2)
```

## Vaccination metadata

### Download link: https://covid19.who.int/who-data/vaccination-metadata.csv

#### Field name	Type	Description
1. ISO3	String	ISO Alpha-3 country code
1. VACCINE_NAME	String	Combined short name of vaccine: “Company - Product name” (see below)
1. PRODUCT_NAME	String	Name or label of vaccine product, or type of vaccine (if unnamed).
1. COMPANY_NAME	String	Marketing authorization holder of vaccine product.
1. FIRST_VACCINE_DATE	Date	Date of first vaccinations. Equivalent to start/launch date of the first vaccine administered in a country.
1. AUTHORIZATION_DATE	Date	Date vaccine product was authorised for use in the country, territory, area.
1. START_DATE	Date	Start/launch date of vaccination with vaccine type (excludes vaccinations during clinical trials).
1. END_DATE	Date	End date of vaccine rollout
1. COMMENT	String	Comments related to vaccine rollout
1. DATA_SOURCE	String	Indicates data source - REPORTING: Data reported by Member States, or sourced from official reports - OWID: Data sourced from Our World in Data: https://ourworldindata.org/covid-vaccinations

```python
# read url into df
vaccination_metadata = pd.read_csv('https://covid19.who.int/who-data/vaccination-metadata.csv')
# save df
vaccination_metadata.to_csv('mined_csv_data/vaccination_metadata.csv')
# inspect df
vaccination_metadata.head(2)
```

## Public Health Social Measures

### Download link: https://covid19.who.int/who-data/phsm-severity-data.csv

#### Field name	Type	Description
1. DATE_START	Date	Date when the PHSM started, changed or extended.
1. COUNTRY	String	Country, territory, area
1. ISO3	String	ISO Alpha-3 country code
1. WHO_REGION	String	WHO regional offices: WHO Member States are grouped into six WHO regions: Regional Office for Africa (AFRO), Regional Office for the Americas (AMRO), Regional Office for South-East Asia (SEARO), Regional Office for Europe (EURO), Regional Office for the Eastern Mediterranean (EMRO), and Regional Office for the Western Pacific (WPRO).
1. MASKS	Integer	Wearing of Masks or facial coverings (see Data Sources section for more details)
1. TRAVEL	Integer	Restrictions on international travel
1. GATHERINGS	Integer	Restrictions on gatherings
1. SCHOOLS	Integer	Adapting or closing schools
1. BUSINESSES	Integer	Adapting or closing offices, businesses, institutions and operations
1. MOVEMENTS	Integer	Restrictions on domestic movement
1. GLOBAL_INDEX	Integer	PHSM Severity Index score – average of all above six indicators
1. MEASURES_IN_PLACE	String	List of individual measures in text

```python
# read url into df
phsm_severity_data = pd.read_csv('https://covid19.who.int/who-data/phsm-severity-data.csv')
# save df
phsm_severity_data.to_csv('mined_csv_data/phsm_severity_data.csv')
# inspect df
phsm_severity_data.head(2)
```
